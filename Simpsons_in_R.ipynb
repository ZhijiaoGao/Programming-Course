```r
# Read data in

simpsons_characters = read.csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-02-04/simpsons_characters.csv')
simpsons_episodes = read.csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-02-04/simpsons_episodes.csv')
simpsons_locations = read.csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-02-04/simpsons_locations.csv')
simpsons_script_lines = read.csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-02-04/simpsons_script_lines.csv')
```

```r
### here we run linear regression to predict imdb rating from viewers

#
#simpsons_episodes = simpsons_episodes.dropna(subset=['imdb_rating', 'us_viewers_in_millions'],)
#change for R syntax
simpsons_episodes <- subset(simpsons_episodes,!is.na(simpsons_episodes$imdb_rating) & !is.na(simpsons_episodes$us_viewers_in_millions))



# create LM
mod <- lm(imdb_rating ~ us_viewers_in_millions, data=simpsons_episodes)


# 4. Get the intercept and coefficient (slope)
summary(mod)


```

```r
summary(mod$coefficient[intercept])
```

```r
#how can I extract only coefficients
intercept = summary(mod)$coefficients[1,1]
coefficient = summary(mod)$coefficients[2,1]
```

```r

# 5. Make predictions
library(ggplot2)

simpsons_episodes$imdb_pred = predict(mod)

#plot_equation = "IMDB rating = " + str(intercept) + " + " + str(coefficient) + "*viewer in millions"

# 6. Visualize the results
ggplot(simpsons_episodes,aes(x=us_viewers_in_millions,y=imdb_rating)) + geom_point(color="blue") + geom_smooth(aes(x=us_viewers_in_millions,y=imdb_pred),method="lm",color="red")


## y = 2.06 + X*0.508
```

```r
library(tidyverse)
frogID_data <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-09-02/frogID_data.csv')
frog_names <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-09-02/frog_names.csv')
```

```r
#merge
frog <- frogID_data %>% left_join(frog_names, by = c("scientificName"))
```

```r
#count
species_counts <- frog %>%
  group_by(scientificName, commonName, subfamily) %>%
  summarise(n_obs = n(), .groups = "drop") %>%
  arrange(desc(n_obs))
```

```r
#RANK TOP 20
top20_species <- species_counts %>% slice_max(n_obs, n = 20)
#PLOT
ggplot(top20_species, aes(x = reorder(commonName, n_obs), y = n_obs, fill = subfamily)) + geom_col() + coord_flip() + labs(
    title = "Top 20 Frog Species by Number of Observations",
    x = "Species",
    y = "Number of Observations",
    fill = "Subfamily"
  ) +
  theme_minimal()
```

```r
ggplot(frog, aes(x = decimalLongitude, y = decimalLatitude, color = subfamily)) +
  geom_point(alpha = 0.6, size = 1) +
  labs(
    title = "Geographic Distribution of Frog Observations",
    x = "Longitude",
    y = "Latitude",
    color = "Subfamily"
  ) +
  theme_minimal()
```
