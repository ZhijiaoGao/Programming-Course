```python
import pandas as pd
pokemon_df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-04-01/pokemon_df.csv')
import matplotlib.pyplot as plt
```

```python
avg_attack = pokemon_df.groupby("type_1")["attack"].mean().sort_values()
```

```python
plt.figure(figsize=(10,6))
avg_attack.plot(kind="barh", color="skyblue")

plt.title("Average Attack by Primary Pokémon Type")
plt.xlabel("Average Attack")
plt.ylabel("Primary Type")
plt.tight_layout()
plt.show()
```

```python
avg_attack = pokemon_df.groupby("egg_group_1")["speed"].mean().sort_values()
```

```python
plt.figure(figsize=(10,6))
avg_attack.plot(kind="barh", color="gold")

plt.title("Average Speed by Pokémon Primary Egg Group")
plt.xlabel("Average Speed")
plt.ylabel("Primary Egg Group")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))

sns.regplot(
    data=pokemon_df,
    x="weight",
    y="attack",
    scatter_kws={"alpha": 0.5},
    line_kws={"color": "red"}
)

plt.title("Relationship Between Pokémon Weight and Attack")
plt.xlabel("Weight")
plt.ylabel("Attack")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.regplot(
    data=pokemon_df,
    x="height",
    y="defense",
    scatter_kws={"alpha": 0.5},
    line_kws={"color": "purple"}
)

plt.title("Relationship Between Pokémon Height and Defense")
plt.xlabel("Height")
plt.ylabel("Defense")
plt.tight_layout()
plt.show()
```
