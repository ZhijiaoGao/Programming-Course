```r


# Load required packages
library(dplyr)
library(readr)
#library(scale)

# Load the nutrient data
# Load in our choice and demographic data

data=read.csv("https://raw.githubusercontent.com/CaitlinLloyd/Psychology_Programming2025/refs/heads/main/Data/demo.csv")
choices=read.csv("https://raw.githubusercontent.com/CaitlinLloyd/Psychology_Programming2025/refs/heads/main/Data/FCT_df.csv")
structure(data)
structure(choices)
```

```r
## we have two different ways in which we can analyse our data - summary statistics or mixed effects

## For example, lets say we want to compare preference for high-fat foods in AN and HC groups

## How can we do this?

```

```r
choices
```

```r
# One way is to summarize preference ratings across high and low-fat foods, and to conduct a t-test or ANOVA comparing groups
long <- choices %>% group_by(ID,fat) %>% summarise_at("rating",mean,na.rm=TRUE)
choices_wide <- long %>% tidyr::pivot_wider(1,values_from=3,names_from=2)
long
choices_wide
```

```r
all <- merge(data,choices_wide,by="ID")
all_long <- merge(data,long,by="ID")
```

```r
## let's try this below - here perform t-test
t.test(high ~ Patient, all)
t.test(low ~ Patient, all)
```

```r
long
```

```r
res<- aov(rating ~
                           Patient * fat +
                           Error(ID/(fat)),data=all_long)
```

```r
summary(res)
```

```r
all_long
```

```r
## how can we plot this data
library(ggplot2)
ggplot(all_long, aes(x = Patient, y = rating, color = fat)) + geom_boxplot() +geom_jitter()

mean_ratings <- all_long %>% group_by(fat, Patient) %>% summarise(mean_rating = mean(rating, na.rm = TRUE), .groups = "drop")
ggplot(mean_ratings, aes(x = Patient, y = mean_rating, fill = fat)) + geom_col(position = position_dodge())
```

```r
install.packages("lmerTest")
library(lmerTest)
lm <- lmer(rating ~ Patient * fat + (1|ID),data=all_long)
```

```r
summary(lm)
```

```r
#The second way is to ask directly whether across all high and low fat trials there is a difference in preference between patient and control
# In the first method, the formula is preference ~ dx; or preference ~ dx*fat + (1|SubID) - we have one or two values per person

# in the second method, the formula is preference ~ fat*dx + (1 + fat|subID)

# here we allow the intercept and slope to vary per person (random effect)
```

```r
## lets say we want to use all of the data
```

```r
lm <- lm(rating ~ Patient * fat ,data=choices)
```

```r
summary(lm)
```

```r
library(ggplot2)
choices$fat_num <- ifelse(choices$fat=="high",1,0)
ggplot(choices,aes(x=fat_num,y=rating,group=ID,color=as.factor(ID))) + geom_line()
```

```r
## when to use random effects:
# see this tutorial: https://bookdown.org/steve_midway/DAR/random-effects.html
# when data points are not independent - eg multiple IV and DV data from one person; or when data are clustered in one group (this violates assumption of independence)
# also violates homogeneity (variance is not the same across grouping structures)
# when you want to generalize to a population but have not collected data from the entire population (eg, individual participants, species)


```

```r
choices <- merge(choices, data,by="ID")
```

```r
lm <- lmer(rating ~ Patient * fat + (1 + fat |ID),data=choices)
```

```r
summary(lm)
```
